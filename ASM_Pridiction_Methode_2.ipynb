```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import arff
from io import BytesIO
```

```python
# Read file
data = arff.loadarff('Autism-Child-Data.arff')
df_child = pd.DataFrame(data[0])
df_child.head()
```

```python
df_child = df_child.drop(['ethnicity','contry_of_res'], axis=1)
df_child.head()
```

```python
df_child.columns.values.tolist()
```

```python
for key in df_child.columns.values.tolist():
    if (type(df_child[key][0])==bytes):
        df_child[key] = df_child[key].str.decode('utf-8')
df_child.head()
```
